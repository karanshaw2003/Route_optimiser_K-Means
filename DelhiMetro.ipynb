```python
def GenerateX(x1 , y1 , x2 , y2):
    X = []
    for i in range(x1.shape[0]):
        new = np.array([x1[i],y1[i]])
        X.append(new)
        
    for i in range(x1.shape[0]):
        new = np.array([x1[i],y1[i]])
        X.append(new)
        
    for i in range(x2.shape[0]):
        new = np.array([x2[i],y2[i]])
        X.append(new)
        
    X = np.array(X)
    return X

X = GenerateX(glongitude , glatitude , ylongitude , ylatitude)
#print(X)
print(X.shape)
```

```python
k = 8
color = ["green","red","blue","yellow","orange","pink","brown","cyan"] 
    

clusters = {}    
    
# Initialise random centres
for i in range(k):
    center = 10* ( (2*np.random.random((2,)) -1) )
    points = []
    
    curCluster = {
        'center' : center ,
        'points' : points ,
        'color' : color[i]
    }
    
    clusters[i] = curCluster 
    
    
    
```

```python
print(clusters)
```

```python
#AllPoints = []

def distance(v1,v2):
    return np.sqrt(np.sum(v1-v2)**2)

# E step.
def AssignPointsToClusters(clusters):
    for ix in range(X.shape[0]):
        curx = X[ix]
        dist=[]

        for i in range(k):
            curCluster = clusters[i]
            curCenter = np.array(curCluster['center'])
            curdist = distance(curx , curCenter)
            dist.append(curdist)

        dist = np.array(dist)
        clusterID = np.argmin(dist)
        curCluster = clusters[clusterID]
        pts = curCluster['points']
        pts.append(curx)
        
        
        
# M step.
def updateClusters(clusters):
    
    for i in range(k):
        curCluster = clusters[i]
        pts = curCluster['points']
        pts = np.array(pts)
        
        if (pts.shape[0]>0):
            newCenter = pts.mean(axis=0)
            clusters[i]['center'] = newCenter
            clusters[i]['points'] = []
            #AllPoints[i] = pts
        
    
    
    
def plotClusters(clusters):
    
    for kx in range(k):
        
        curCluster = clusters[kx]
        pts = curCluster['points']
        pts = np.array(pts)
        curColor = curCluster['color']
        #Plot the points
        try:
            plt.scatter(pts[:,0],pts[:,1],c=curColor)
        except:
            pass
        

        #Plot the cluster centers
        curCenter = curCluster['center']
        plt.scatter(curCenter[0] , curCenter[1],color='k',marker="^") 
        #plt.legend()
    plt.show()


```

```python
AssignPointsToClusters(clusters)
#print(clusters)
plotClusters(clusters)
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum(v1-v2)**2)


def model(x,k):
    
    sk_kmeans = KMeans(k)
    sk_kmeans.fit(x)
    return sk_kmeans

def PlotFig(x,sk_kmeans):
    plt.figure()
    plt.scatter(0,0,c='r',marker="*",s=400)
    i=0
    
    for point,label in zip(x,sk_kmeans.labels_):
        plt.scatter(point[0],point[1],color = colors[label])

    for center in sk_kmeans.cluster_centers_:
        plt.scatter(center[0] , center[1] , color='k' , marker="^" , s=100)
        
    plt.show()
    
    
    for point,label in zip(x,sk_kmeans.labels_):
        plt.scatter(point[0],point[1],color = 'b')
    for center in sk_kmeans.cluster_centers_:
        plt.scatter(center[0] , center[1] , color='k' , marker="^" , s=100)
        
    plt.show()


def KmeansCluster(x,k):
    
    sk_kmeans = model(x,k)
    
    myCluster = sk_kmeans.labels_
    
    #points in every cluster
    pts = []
    for i in range(k):
        pts.append([])
    for i in range(x.shape[0]):
        curCluster = myCluster[i]
        pts[curCluster].append(x[i])
        
    
    clusterCenters = sk_kmeans.cluster_centers_
    flag = True 
    dist = []
    for i in range(k):
        cur_distances = []
        curCenter = clusterCenters[i]
        curpts = pts[i]
        for j in range(len(curpts)):
            d = distance(np.array(curCenter) , np.array(curpts[j]))
            if(d>5):
                #global flag
                flag = False
                break
            cur_distances.append(d)
        
        
    return flag , sk_kmeans
    
def GetBusStops(x):
    k=10
    while(True):
        flag , sk_kmeans = KmeansCluster(x,k)
        if(flag==True):
            #correct model
            
            #plot the model
            PlotFig(x,sk_kmeans)
            
            break
        else:
            k+=1
            continue
        
    return sk_kmeans
            
    
    
```
